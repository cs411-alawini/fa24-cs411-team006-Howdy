# Dataset 1

```python
import pandas as pd
import numpy as np
file_path = 'Sleep_health_and_lifestyle_dataset.csv'
sleep_data = pd.read_csv(file_path)
sleep_data.head()
```

```python
# drop person id
sleep_data = sleep_data.drop(columns=['Person ID'])
sleep_data.head()
```

```python
sleep_data['Gender'] = sleep_data['Gender'].map({'Male': 0, 'Female': 1})
sleep_data['Occupation'] = sleep_data['Occupation'].astype('category').cat.codes
sleep_data['BMI Category'] = sleep_data['BMI Category'].astype('category').cat.codes
sleep_data['Sleep Disorder'] = sleep_data['Sleep Disorder'].apply(lambda x: 0 if x == 'None' else 1)
sleep_data[['Systolic Pressure', 'Diastolic Pressure']] = sleep_data['Blood Pressure'].str.split('/', expand=True).astype(float)
sleep_data = sleep_data.drop(columns=['Blood Pressure'])
sleep_data
```

```python
correlation_matrix = sleep_data.corr(numeric_only=True)
correlation_matrix
```

sleep duration - stress level - Quality of Sleep - heart rate

occupation - BMI Category - systolic pressure - disatolic pressure

physical activity level - daily steps

things with weak relationship with sleep disorder: gender, physical activity level stress level, daily steps
